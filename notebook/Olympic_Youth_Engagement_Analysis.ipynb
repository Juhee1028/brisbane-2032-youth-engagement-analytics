# 1. Introduction

The Olympic Games, as a global sporting event, can generate significant economic and social benefits, including increased tourism, industry development, and enhanced city reputation as long-term impacts. In addition, substantial short-term revenue can be generated through ticket sales, sponsorships, licensing agreements, and broadcasting rights (Chang et al., 2025). To maximise these benefits, the Australian Sports Commission (ASC) should strengthen long-term Olympic engagement through the development of a sustainable sports fanbase.

Fanbase development is closely associated with sports interest and participation, particularly among children and adolescents who represent the next generation of Olympic audiences, participants, and supporters. Young people who actively participate in sports are more likely to maintain long-term interest in sporting events, follow athletes, and engage with Olympic-related content in the future. Therefore, understanding youth participation patterns is important to identify potential target groups for future Olympic engagement strategies toward the Brisbane 2032 Olympic Games.

Different participation groups may also demonstrate different levels of participation continuity, gender balance, and motivational characteristics. Identifying these differences may help the ASC better understand which groups have stronger potential for long-term Olympic engagement and which participation factors should be considered when developing future engagement approaches.

Furthermore, social media also plays an increasingly significant role in shaping sports engagement among young people. Through real-time updates, interactive communication, and personalised content, social media can strengthen audiences’ emotional connection to sporting events and athletes (Lau, n.d.). Although Australia has introduced restrictions on social media accounts for people under 16, many young people continue to actively use social media platforms (Reuters, 2026). Therefore, social media may provide opportunities to support future Olympic engagement strategies among young people.

In addition, effective engagement strategies also require an understanding of the motivations related to sports participation. Motivation factors may influence how young people engage with sports activities, athletes, and Olympic-related content. Understanding these motivational differences may help support more targeted engagement approaches across different participation groups.

Therefore, this analysis investigates:
- Current participation patterns among children and adolescents to identify potential target groups for Olympic fanbase development,
- Participation clusters associated with participation level, adolescent continuity, and gender balance,
- The extent to which social media is currently associated with Olympic and sports engagement, and
- How motivation factors may inform future Olympic engagement approaches across different participation groups.

This analysis aims to address the following main question:

**"How can youth participation patterns and motivation factors be used for future Olympic engagement strategies toward the Brisbane 2032 Olympic Games?"**

## 2.2. Data

```python
import pandas as pd
import plotly.express as px

df = pd.read_excel(
    './data_Ass2/AusPlay-By-Sport-Data.xlsx',
    sheet_name = '2',
    nrows =142,
    skiprows = 158,
    header=None,
    names = ['total', 'C','males','females','G','0-4','5-8','9-11','12-14'])

df
```

- Dataset: Australian Sports Commission AusPlay "By Sport Data Tables" in 2026

```python
# mapping olympic sports
olympic_sports = [ 'Archery', 'Artistic swimming', 'Synchronised swimming', 'Athletics', 'Athletics, track and field', 
                   'Badminton', 'Baseball','Basketball', 'Beach volleyball', 'Athletics, track and field', 
                   'Boxing', 'Canoeing','Canoeing/Kayaking', 'Cricket', 'Cycling','Diving', 'Equestrian', 
                   'Fencing', 'Football','Football/soccer', 'Golf', 'Gymnastics', 'Handball','Field hockey','Hockey', 
                   'Judo', 'Lacrosse', 'Modern pentathlon', 'Rowing', 'Rugby', 'Rugby union','Sailing', 'Shooting', 
                   'Shooting/field', 'Shooting/range', 'Skateboarding', 'Skate' ,'Softball', 'Sport climbing', 'Squash',
                   'Surfing', 'Swimming', 'Table tennis', 'Taekwondo', 'Taekwondo/Taekwon-do','Tennis', 'Triathlon', 
                   'Volleyball','Volleyball (indoor and outdoor)', 'Water polo', 'Weightlifting', 'Weight lifting', 
                   'Wrestling']
                   

# filtering
df_olympic = df[df.index.isin(olympic_sports)]

df_olympic
```

- Focused only on sports associated with the Olympic Games
- Some Olympic sport names did not exactly match the sport names used in the AusPlay dataset. Thus, both the official Olympic sport names and the corresponding dataset sport names were included in the filtering list to ensure that all relevant Olympic sports were captured during the filtering process.

**Resources**: Olympic sport list [Link](https://en.wikipedia.org/wiki/Portal:Olympic_Games)

```python
# excluded 0 value in total of dataset
df_olympic = df_olympic[df_olympic['total'] > 0]

# 5-14 age group (children + adolescents)
df_olympic['5-14'] = df_olympic['5-8'] + df_olympic['9-11'] + df_olympic['12-14']

# adolescent age group
df_olympic['9-14'] = df_olympic['9-11'] + df_olympic['12-14']

# replace NaN
df_olympic = df_olympic.fillna(0)

# adolescent ratio
df_olympic['adolescent_ratio'] = df_olympic['9-14'] / df_olympic['5-14']

# gender ratios
df_olympic['male_ratio'] = df_olympic['males'] / df_olympic['total']

df_olympic['female_ratio'] = df_olympic['females'] / df_olympic['total']

# select relevant columns
df_olympic = df_olympic[['total','male_ratio','female_ratio','5-14','5-8','9-11','12-14','adolescent_ratio']]

df_olympic.head()
```

- Extracted variables related to youth's sport participation, adolescents' sport participation, and gender participation patterns

- `5-14`: Overall participation level of children and adolescents in each Olympic sport to identify which sports are currently most popular among young people.

- `9-14`: The number of older children and adolescents participating in each sport. This was used as a proxy indicator of continued participation during adolescence.

- `adolescent_ratio`: The proportion of participants aged 9-14 within the total 5-14 age group. A higher ratio may indicate a stronger retention of participation into adolescence

- `male_ratio`, `female_ratio`: Represents the proportion of male/female participants in each sport to identify male-dominant, female-dominant, and gender-balanced sports.

The `gender_balance_score` was calculated using the absolute difference between `male_ratio` and `female_ratio`.

- lower `gender_balance_score`

### 2.3.3. Target Group Identification (Clustering)

K-means clustering was used to group Olympic sports with similar participation characteristics. This helps move the analysis from individual sports to broader **target groups**. 

The clustering analysis suggests that different Olympic sports may require different engagement and participation strategies rather than a single generalised approach. For example, a sport with high participation but weaker adolescent retention requires a different strategy from a sport with low participation but strong adolescent retention.

```python
from sklearn.preprocessing import minmax_scale
from sklearn.cluster import KMeans

# Use only Olympic sports with youth participation
cluster_df = df_olympic[df_olympic['5-14'] > 0].copy()

# Scale variables
cluster_df['participation_n'] = minmax_scale(cluster_df['5-14'])

cluster_df['adolescent_n'] = minmax_scale(cluster_df['adolescent_ratio'])

cluster_df['gender_balance_n'] = minmax_scale(cluster_df['gender_balance_score'])

cluster_df.head()
```

The clustering was based on three normalised variables:

- `participation_n`: overall youth participation among children aged 5–14
- `adolescent_n`: adolescent representation within youth participation
- `gender_balance_n`: gender balance pattern between male and female participants

```python
# K-means clustering (2 clusters)

km2 = KMeans(n_clusters=2,random_state=0,n_init=10).fit(cluster_df[['participation_n','adolescent_n','gender_balance_n']])

cluster_df['km2'] = km2.labels_

# K-means clustering (3 clusters)

km3 = KMeans(n_clusters=3,random_state=0,n_init=10).fit(cluster_df[['participation_n','adolescent_n','gender_balance_n']])

cluster_df['km3'] = km3.labels_

#cluster_df.head()

cluster_df[['participation_n','adolescent_n','gender_balance_n','km2','km3']].head()
```

This table shows the normalised values used for clustering and the assigned cluster labels.

- `km2`: result from the 2-cluster model
- `km3`: result from the 3-cluster model

The 3-cluster model was used because it provides a more detailed grouping of sports based on participation, adolescent representation, and gender balance.

**Key Findings**

- Each sport was assigned to a cluster based on similar participation characteristics.
- The clustering identifies sports with similar behavioural patterns rather than ranking sports from best to worst.
- Sports with similar participation patterns may require similar engagement and promotional strategies.


```python
cluster_df['km3'] = cluster_df['km3'].astype(str)

fig = px.scatter_3d(
    cluster_df,
    x='participation_n',
    y='adolescent_n',
    z='gender_balance_n',
    color='km3',
    hover_name=cluster_df.index,
    title='K-means Clustering of Sports Participation',
    labels={
        'participation_n': 'Normalised Participation',
        'adolescent_n': 'Normalised Adolescent Ratio',
        'gender_balance_n': 'Normalised gender balance'
    },
    width= 700,
    height = 700 
)

fig.show()
```

This 3D scatter plot visualises the clustering results using:

- normalised participation,
- normalised adolescent representation,
- and normalised gender balance.

Each point represents one Olympic sport, while colours represent cluster groups.

**Key Findings**

- The three clusters are visually separated, indicating that the clustering model identified different participation patterns across sports.
- **Cluster 0** appears around low participation but higher gender balance.
- **Cluster 1** appears around low participation but strong adolescent representation.

#### Ethical Considerations

**Simplification of Complex Behaviours**

K-means clustering simplifies sports into grouped categories based on numerical similarities.

While this helps identify broader participation patterns, clustering may overlook contextual differences between sports, cultural meaning, participation motivations, and accessibility challenges.

For example, two sports may appear statistically similar while having completely different social or economic participation barriers.

Therefore, cluster labels should not be interpreted as fixed or definitive categories.

---
**Risk of Strategic Exclusion**

The clustering analysis may unintentionally prioritise sports with higher participation or stronger adolescent retention.

This creates a potential ethical risk because:

- lower participation sports may receive less visibility,
- niche sports may receive reduced investment,

```python
# get data from server
server_response = requests.get(full_url)
server_data = server_response.json()  #JSON format
resp_data = server_data.get('response','')
if resp_data == '':
    print("ERROR obtaining results:",server_data)
else:
    print("SUCCESS!")
    print(f"{resp_data['total']} results found available in {resp_data['pages']} pages")
    print(f"{resp_data['pageSize']} results per page")
    results = resp_data.get('results',[])
```

```python
results[0] # to check the structure of article
```

```python
def articles_from_page_results(page_results): # to extract the publication date, title, and body text from each article.
    articles = {}
    for result in page_results:
        article_date = result['webPublicationDate']
        article_title = result['webTitle']+f" [{article_date}]"
        article_html = result['fields']['body']
        article_text = re.sub(r'<.*?>','',article_html)
        articles[article_title] = article_text
    return articles

def get_all_articles_for_response(response_json,full_url): # 
    total_pages = response_json['pages']
    total_articles = response_json['total']
    print(f"Fetching {total_articles} articles from {total_pages} pages...")
    all_articles = {}
    page1_articles = articles_from_page_results(response_json['results'])
    all_articles.update(page1_articles)
    print("Added articles for page: 1")
    
    for page in range(2,total_pages+1): # to retrieve articles from every page, because Guardian API returned only 10 articles per page
        print("Getting articles from API for page:",page)
        page_response = requests.get(full_url+f"&page={page}")
        page_data = page_response.json()['response']
        print("Processing results for page:",page_data['currentPage'])
        page_articles = articles_from_page_results(page_data['results'])
        print(f"Fetched {len(page_articles)} articles.")
        all_articles.update(page_articles)
        print("Added articles for page:",page)
        print(f"Status: {len(all_articles)} articles.")
        time.sleep(1) # to avoid sending too many requests
    
    print(f"FINISHED: Fetched {len(all_articles)} articles.")
    return all_articles
```

#### (1) Process for generating dataframe for term analysis

```python
# Create a dataframe to hold top terms for each analysis type

terms_df = pd.DataFrame(index=articles_filtered.keys(),columns=['count', 'tfidf', 'nmf'])
terms_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

# Count (term frequency)
# Set parameters appropriate to your data
count_vectorizer = CountVectorizer(max_df=0.8,min_df=1,max_features=1000,stop_words="english")

# Get the document vectors
count_dt_matrix = count_vectorizer.fit_transform(articles_filtered.values())

# Get the terms identified during the vectorization process
feature_names = count_vectorizer.get_feature_names_out()

# create a new dataframe with the matrix
count_df = pd.DataFrame(count_dt_matrix.toarray(), index=list(articles_filtered.keys()) ,columns=feature_names)

# for each doc, get the 10 columns with the Largest counts
for idx in terms_df.index:
    counts = dict(count_df.loc[idx].sort_values(ascending=False).head(10))
    #print(counts)
    terms_df.at[idx,'count'] = list(counts.keys()) # Just the list of terms


# tfidf
# Set parameters appropriate to your data
tfidf_vectorizer = TfidfVectorizer(
    max_df=0.8, min_df=2, max_features=1000, stop_words="english"
)
# Get the document vectors
tfidf_dt_matrix = tfidf_vectorizer.fit_transform(articles_filtered.values())

# list of feature names
feature_names = tfidf_vectorizer.get_feature_names_out()

# create a df to combine matrix with feature names
tfidf_df = pd.DataFrame(tfidf_dt_matrix.toarray(), index=list(articles_filtered.keys()) ,columns=feature_names)

for idx in terms_df.index:
    tfidf = dict(tfidf_df.loc[idx].sort_values(ascending=False).head(10))
    terms_df.at[idx,'tfidf'] = list(tfidf.keys()) 

terms_df
```

```python
#NMF topic modelling
# identify broader topics across Olympic/sport articles

from sklearn.decomposition import NMF

# Set the number of topics
num_topics = 15

# Create the model
nmf_model = NMF(n_components=num_topics,init='random',beta_loss='frobenius')

# Fit the model to the data and use it to transform the data
doc_topic_nmf = nmf_model.fit_transform(tfidf_dt_matrix)

topic_term_nmf = nmf_model.components_

# Get the topics and their terms
nmf_topic_dict = {}
for index, topic in enumerate(topic_term_nmf):
    zipped = zip(feature_names, topic)
    top_terms=dict(sorted(zipped, key = lambda t: t[1], reverse=True)[:10])
    #print(top_terms)
    top_terms_list= {key : round(top_terms[key], 4) for key in top_terms.keys()}
    nmf_topic_dict[f"topic_{index}"] = top_terms_list

# Print the topics with their terms    
for k,v in nmf_topic_dict.items():
    print(k)
    print(v)
    print()
```

```python
# assign the strongest NMF topic to each article

for article_name,topic in zip(terms_df.index, doc_topic_nmf):
    topic_num = topic.argmax()
    top_topic = nmf_topic_dict[f"topic_{topic_num}"]
    terms_df.at[article_name,'nmf'] = list(top_topic.keys())

terms_df
```

## 3.5. Ethical Consideration

**Representation and Interpretation of Data**

This analysis should acknowledge that media data does not represent completely objective reality. The selection of keywords, filtering methods, and topic modelling techniques influences which themes become visible within the analysis. Data is never fully “raw” because human decisions shape how information is collected, represented, and interpreted.
- Therefore, the findings should be interpreted as representations of media coverage patterns rather than absolute truths about Olympic engagement or youth behaviour.
---
**Measurement Bias and Limitations**

Keyword frequency analysis may simplify complex social discussions into numerical measurements. High keyword frequency does not necessarily indicate social importance, while low frequency does not automatically mean lack of public interest.

For example:
- the low number of youth and digital keyword mentions does not prove that young people are uninterested in Olympic sports,
- but suggests that these topics were not strongly represented within the selected media articles.

This reflects the ethical concern that “all models are wrong, but some are useful” (Box, 1976), meaning analytical models are approximations rather than perfect representations of reality.

---
**Fairness**

Media coverage may unintentionally prioritise highly successful athletes, popular sports, controversial issues, or commercially attractive events.

As a result, smaller sports, minority groups, or less visible communities may become underrepresented within the dataset.

This creates potential fairness concerns because media visibility can influence the perceived importance of sports, public attention, and future investment or engagement strategies.

Bias may thus emerge through both media reporting practices and analytical decisions made during filtering and modelling processes.

---
**Transparency and Accountability**

Transparency is important throughout this analysis process. The selected search queries, filtering criteria, keywords, and text mining methods should be clearly documented so that the analysis process can be understood and critically evaluated.

In addition, topic modelling techniques such as NMF reduce large amounts of text into simplified mathematical patterns. While useful for identifying themes, these models cannot fully capture social context, emotion, nuance, or human meaning. Therefore, interpretation should remain cautious and critically reflective.

    - **Females** showed comparatively higher motivation values for physical health (0.69), mental health (0.42), outdoor or nature-related reasons (0.54), and weight loss (0.56) than males. This may suggest that female participation is more strongly associated with *wellbeing, lifestyle*, and *health-related motivations*.

    - **Males** demonstrated relatively higher motivation values for hobby-related participation (0.32) and training or competition-oriented motivations (0.25) than females. This may indicate that male participation is more associated with *sport-specific interest, training*, and *performance-related engagement*.

- Social-related motivations appeared relatively similar across both groups, suggesting that social interaction and community engagement are important participation factors regardless of gender.

**Insights**

- Participation motivations are not evenly distributed across male and female groups. Although several motivations, such as physical health and enjoyment, appeared important across both groups, some noticeable differences were identified in wellbeing-, lifestyle, and competition-related motivations.

These findings may suggest that sports with stronger gender-dominant participation patterns could require more targeted engagement approaches for underrepresented genders rather than relying only on general participation promotion strategies.

## 4.4. Insights

The motivation analysis suggests that different participation approaches may be required depending on the participation characteristics identified in the previous clustering analysis.

- For sport groups with lower overall participation levels, the overall motivation analysis suggests that **broader participation-focused approaches** may become more effective when emphasising physical health, enjoyment, outdoor activity, and social engagement rather than focusing only on elite competition or Olympic performance.

- The gender-related motivation analysis suggests that sports with stronger gender-dominant participation patterns may require **more targeted engagement approaches** for underrepresented genders.
    - For example, sports with lower female participation may benefit from emphasising health, wellbeing, outdoor activity, and social engagement factors. In contrast, sports with lower male participation may benefit from approaches associated with hobby participation, training, and competition-related engagement.

Overall, these findings suggest that future Olympic engagement strategies may become more effective when they reflect the different participation motivations associated with different participation groups rather than relying only on broad generalised promotion approaches.

## 4.5. Ethical Considerations

This motivation analysis uses aggregated participation data rather than individual-level personal information, which helps reduce direct privacy concerns. However, several ethical considerations should still be acknowledged when interpreting and applying the findings.

1. Motivation patterns should not be interpreted as fixed characteristics of all individuals within a gender group. The findings only represent broad participation trends from the survey data and may not reflect the motivations or experiences of every participant.

2. There is a risk that gender-related motivation differences may reinforce stereotypes if interpreted too generally. For example, higher health- or competition-related motivations observed within one gender group should not be used to assume that all individuals within that group share the same participation interests or behaviours.

3. Motivation-based engagement approaches may unintentionally prioritise certain participation groups while overlooking others. Strategies developed from these findings should therefore aim to support inclusiveness and broader participation opportunities rather than strengthen existing gender-dominant participation patterns.

4. The analysis is based on self-reported survey responses, meaning that participation motivations may be influenced by subjective interpretation, response bias, or changing personal circumstances.

    - Therefore, the findings should be interpreted as general participation tendencies rather than precise representations of all youth sport participants.

# 5. Conclusion

This analysis investigated how youth participation patterns and participation motivations may inform future Olympic engagement strategies toward the Brisbane 2032 Olympic Games.

The participation and clustering analysis showed that different Olympic sport groups demonstrate different participation characteristics related to participation level, adolescent retention, and gender balance. Several sport groups showed low overall participation, while others demonstrated stronger gender-dominant participation patterns or weaker adolescent participation continuity. These findings suggest that different sports may require different types of engagement support rather than relying on a single generalised strategy.

The media analysis further examined whether Olympic and sports-related media coverage is currently associated with youth and digital engagement topics. These findings showed that social media and youth-oriented engagement concepts currently have relatively limited visibility within Olympic and sports-related media coverage. Most articles focused more heavily on athlete performance, competitions, and major sporting events rather than long-term youth engagement. This may suggest that social media is not yet being strongly utilised as a tool for long-term Olympic youth engagement despite its potential relevance for younger audiences.

Based on these findings, the motivation analysis explored which participation motivations may be important if future Olympic engagement strategies increasingly utilise social media and digital engagement approaches. The overall motivation analysis showed that sport participation is primarily associated with physical health, enjoyment, lifestyle-related benefits, outdoor activity, and social engagement rather than elite competition or professional ambition alone.

- The gender-related motivation analysis also suggested that participation motivations are not evenly distributed across male and female groups. This result may suggest that sports with stronger gender-dominant participation patterns could require more targeted engagement approaches for underrepresented genders rather than relying only on broad participation promotion strategies.

Overall, the findings suggest that future Olympic engagement strategies may become more effective when they reflect different participation characteristics and participation motivations across youth participation groups while more actively utilising social media as a long-term engagement tool toward Brisbane 2032.
